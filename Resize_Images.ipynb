# Resize images to fit into a CNN

```python
import matplotlib.pylab as plt
import urllib.request
import pandas as pd
import numpy as np
import io
import os
```

```python
path = 'Images'
files = [os.path.join(path, file) for file in  os.listdir(path)]
```

```python
data = []
for file in files:
    image = plt.imread(file)
    data.append(image.shape)
```

```python
df = pd.DataFrame(data, columns=['x', 'y', 'z'])
```

```python
for col in df.columns:
    df[col].hist()
    plt.title(col)
    plt.show()
```

```python
df['xy'] = df['x'] / df['y']
df['xy'].hist(bins=20)
plt.title('Seitenverh√§ltnisse: 1:x')
plt.show()
```

```python
from PIL import Image
 
def resize_image(input_image_path, output_image_path, size):
    print(input_image_path)
    original_image = Image.open(input_image_path)
    width, height = original_image.size
    print('The original image size is {wide} wide x {height} high'.format(wide=width, height=height))
    original_image.show()
    
    resized_image = original_image.resize(size)
    width, height = resized_image.size
    print('The resized image size is {wide} wide x {height} high'.format(wide=width, height=height))
    resized_image.show()
    #resized_image.save(output_image_path)
    
    
resize_image(input_image_path=files[1], output_image_path='', size=(150, 100))
```
